```python
#Add the filtered data in dff

dff = df[(df['Event distance/length'].isin(['50Km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
dff.shape
```

```python
#Remove USA from event name
dff['Event name'] = dff['Event name'].str.split('(').str.get(0)
```

```python
dff.head()
```

```python
#Create correct Age column

dff['Athlete age'] = 2020 - dff['Athlete year of birth']
```

```python
#Clean up h from Athlete performance

dff['Athlete performance'] = dff['Athlete performance'].str.replace('h','')
```

```python
dff.head(5)
```

```python
#Drop Columns Athlete club, Athlete age category

dff.drop(columns=['Athlete club','Athlete age category' ], inplace=True)
```

```python
dff.head()
```

```python
#Show total null values in all columns

dff.isna().sum()
```

```python
#Drop null rows

dff = dff.dropna()
```

```python
dff.shape
```

```python
#Check for duplicates

dff[dff.duplicated() == True]
```

```python
#Reset index

dff.reset_index(drop = True)
```

```python
#Fix data types

dff.dtypes
```

```python
dff['Athlete age'] = dff['Athlete age'].astype(int)
```

```python
dff['Athlete average speed'] = dff['Athlete average speed'].astype(float)
```

```python
dff.head()
```

```python
#Rename Columns

#Year of event                  
#Event dates                   
#Event name                    
#Event distance/length         
#Event number of finishers      
#Athlete performance           
#Athlete country              
#Athlete year of birth       
#Athlete gender                
#Athlete average speed         
#Athlete ID                   
#Athlete age                  
```

```python
dff = dff.rename(columns = {
    'Year of event': 'year',                
    'Event dates': 'race_day',                
    'Event name': 'race_name',                 
    'Event distance/length': 'race_length',        
    'Event number of finishers': 'race_number_of_finishers',     
    'Athlete performance': 'athlete_performance',        
    'Athlete country': 'athlete_country',  # Corrected column name
    'Athlete year of birth': 'athlete_birth_year',     
    'Athlete gender': 'athlete_gender',             
    'Athlete average speed': 'athlete_average_speed',       
    'Athlete ID': 'athlete_id',                  
    'Athlete age': 'athlete_age'
})
```

```python
# Drop the column 'athlete_gender' to fix the mistake
dff = dff.drop(columns=['athlete_gender'])
```

```python
dff.head(5)
```

```python
#charts and graphs

sns.histplot(dff['race_length'])
```

```python
# Merge df and dff on the common key columns 'athlete_id' and 'Athlete ID'
dff = pd.merge(dff, df[['Athlete ID', 'Athlete gender']], left_on='athlete_id', right_on='Athlete ID', how='left')

# Drop the duplicate key column 'Athlete ID' after merging
dff.drop(columns=['Athlete ID'], inplace=True)
```

```python
#Rename Athlete gender

dff.rename(columns = {'Athlete gender' : 'athlete_gender'}, inplace=True)
```

```python
#Plot race_length per gender graph

sns.histplot(dff, x = 'race_length', hue = 'athlete_gender')
```

```python
#plot race length per average speed 

sns.displot(dff[dff['race_length'] == '50mi']['athlete_average_speed'])
```

```python
#plot race length per average speed and gender

sns.violinplot(data = dff, x = 'race_length', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
#plot race average speed per athlete age and hue the gender

sns.lmplot(data = dff, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
#Find the top 10 average speed per age and count how many

dff.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head(10)
```

```python
#Find the slowest top 10 average speed per age and count how many

dff.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19').head(10)
```
